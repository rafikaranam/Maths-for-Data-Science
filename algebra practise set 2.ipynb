# Problem set 2:
Jill design solar panel as hobby,
On April 1, Jill's Mark1 design can produce 1kj/day
on May 1, Jill's Mark2 design can produce 4kj/day
1. What day is it when Jill's Mark2 design produce as much total energy as Mark1
2. What is the total amount of energy produced by that day

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

Step 2: Set the line points for slope (Energy and #days)

```python
d = np.linspace(0,100,1000)
```

Step 3: declare varaibles for energy and days 
        M1 energy per day is 1kj from April 1
        i.e M1_energy = 1d (if it's day1 energey will be 1 * 1 )
        M2 energy per day is 4kj from May 1
        i.e M2_energy = 4(d-30)
            M1_energy = M2_energy as we need to find the number of day it wil take for M2 to produce the same energy   produce by M1

```python
m1_energy = 1*d
```

```python
m2_energy = 4*(d-30)
```

Now let us plot the graph

```python
fig, ax = plt.subplots()
plt.title('# of days where M1 and M2 produce same energy')
plt.xlabel('days')
plt.ylabel('Energy in KJ')
ax.set_xlim([0,100])
ax.set_ylim([0,100])
ax.plot(d, m1_energy, c='green')
ax.plot(d, m2_energy, c='brown')
ax.axvline(x=40, color = 'purple', linestyle='--')
ax.axhline(y=40, color = 'orange', linestyle='--')
```

So it will take 40 days from April 1 to produce 40kj for M1 and 10 days for M2 

Total amount produce is 40+40 = 80 kj by that 40th day i.e May 10th
