# Problem set 1:

A bank robber started running at 150km/hr and A police officer started 5 minutes late to catch robber with a speed of 180km/hr. Find the distance and time when robber and police meet.

step 1: Import Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

Step 2: Set the X axis for time interval

```python
t = np.linspace(0,40,1000)
```

Step 3: declare varaibles for robber and police 
        Here the distance travelled is same for robber and police so we will equate both of their distances to solve 
        proble
        distance_robber = 2.5km/hr * t
        distance_police = 3*(t-5)

```python
d_r = 2.5*t
```

```python
d_p = 3*(t-5)
```

Now let us plot the graph

```python
fig, ax = plt.subplots()
plt.title('Robber caught')
plt.xlabel('Time (in mins)')
plt.ylabel('Distance (in km)')
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, d_r, c='green')
ax.plot(t, d_p, c='brown')
plt.axvline(x=30, color = 'purple', linestyle='--')
plt.axhline(y=75, color = 'orange', linestyle='--')
```
