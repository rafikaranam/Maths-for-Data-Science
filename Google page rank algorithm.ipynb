```python
import numpy as np
import numpy.linalg as la
```

#### Create Link Matrix

```python
L = np.array([[0,1/2,0,0],[1/3,0,0,1/2],[1/3,0,0,1/2],[1/3,1/2,1,0]])
L
```

```python
#Lets create initial matrix
```

```python
r = 100*np.ones(4)/4
```

```python
#Loop until we get the required tolerance
```

```python
lastR = r
r = L @ lastR 
i = 0
while la.norm(lastR-r)>0.01:
    lastR = r
    r = L @ lastR
    i=i+1
    print(r)
print(f"{i} number of iterations required for convergenc")
```
