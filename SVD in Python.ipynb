```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#define plot style 
sns.set()
%pylab inline
pylab.rcParams['figure.figsize'] = (8,8)
np.set_printoptions(suppress = True)
```

```python
def plotVec(vec, col, alpha=1):
    """
    vec - array vector co ordinates
    col - array color
    alpha - float opacity
    """
    plt.axvline(x=0, color = '#A9A9A9', zorder = 0)
    plt.axhline(y=0, color = '#A9A9A9', zorder = 0)
    

    for i in range(len(vec)):
        if(isinstance(alpha,list)):
            alpha_i = alpha[i]
        else:
            alpha_i = alpha
        x = np.concatenate([[0,0],vec[i]])
        

        plt.quiver([x[0]],
                  [x[1]],
                  [x[2]],
                  [x[3]], 
                  angles = 'xy', scale_units='xy',scale =1, color = col[i],
                  alpha=alpha_i)
        
    
    
```

```python
green = '#006400'
red = '#FF0000'
v = [1,2]
plotVec([v],col=[green],alpha=1)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.show()
```

```python
#scaling 
s = np.array([[2,0],[0,2]])
b = s@v
plotVec([v],col=[green],alpha=1)
plotVec([b],col=[red],alpha=1)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.show()
```

```python
#Rotation by 90 degrees anti clockwise
r = np.array([[0,-1],[1,0]])
c = r@v
plotVec([v],col=[green],alpha=1)
plotVec([c],col=[red],alpha=1)
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.show()
```

```python
#Rotation and stretching
A= np.array([[2,1],[-1,1]])
B = A@v
plotVec([v],col=[green],alpha=1)
plotVec([B],col=[red],alpha=1)
plt.xlim(-4,4)
plt.ylim(-4,4)
```

```python
""""
Circle equation x^2+y^2=1
y = sqrt(1-x^2)
lets plot unit circle
"""

x = np.linspace(-1,1,1000000)
y = np.sqrt(1-x**2)
plt.plot(x,y,sns.color_palette().as_hex()[1])
plt.plot(x,-y,sns.color_palette().as_hex()[1])
plt.xlim(-2,2)
plt.ylim(-2,2)
```

```python
"""
Modify unit circle and basis vector by applying matrix
"""
def plotmatrix(matrix,vectcol=["#FF0000","#FFA500"]):
    x = np.linspace(-1,1,1000000)
    y = np.sqrt(1-x**2)
    #Modified unit circle
    x1 = matrix[0,0]*x + matrix[0,1]*y
    y1 = matrix[1,0]*x + matrix[1,1]*y
    x1n = matrix[0,0]*x - matrix[0,1]*y
    y1n = matrix[1,0]*x - matrix[1,1]*y
    
    #vectors
    u1 = [matrix[0,0],matrix[1,0]]
    v1 = [matrix[0,1],matrix[1,1]]
    
    plotVec([u1,v1],col = [vectcol[0],vectcol[1]])
    plt.plot(x1,y1,'g',alpha = 0.5)
    plt.plot(x1n,y1n,'g',alpha=0.5)
```

```python
A = np.array([[2,1],[-1,1]])
U = np.array([[1,0],[0,1]])
print('Unit circle')
plotmatrix(U)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
print('Unit circle transformed by A')
plotmatrix(A)
plt.xlim(-3,3)
plt.ylim(-3,3)
plt.show()
```
